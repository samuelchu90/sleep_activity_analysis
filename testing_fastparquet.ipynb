This is also a good notebook to learn how to run SALA commands

```python
#imports
from SALA import processing
from SALA import plots
import pandas as pd
import numpy as np
```

```python
directory = {
    'summer':'data/summer_16_truncated'
}
timezone = "America/Los_Angeles"
latitude = 47.61
longitude = 122.33
```

```python
sala_summer = processing.SALAFrame(latitude, longitude, timezone, directory = directory)
sala_summer
```

```python
summer_raw_data = sala_summer.get_raw_data_from_key('summer')
```

Output raw data to parquet

```python
summer_raw_data.to_parquet("data/summer16truncated_raw.parquet",
                               engine = "fastparquet", compression="gzip")
```

```python
sala_summer.process_data(summer_raw_data, [[5],[50],[500]])
```

```python
sala_summer.data = sala_summer.sun_timings()
```

```python
short_frame, timing_data = sala_summer.process_sleep(summer_raw_data, num_sleeps = 1)
```

To export processed SALA summary

```python
sala_summer.export('data/summer16_processed')
```
